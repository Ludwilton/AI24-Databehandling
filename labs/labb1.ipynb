## Correlation between spending and graduation rate
For this graph i wanted to look at the correlation between total spending* costs and graduation rate and as we can see from the graph and calculated correlation the spending does have a weak to moderate influence on graduation rate. Although we can probably find more significant factors.




--- 

\*room, board and books costs, instructional expenditure and personal spending

```python


college.loc[college['Grad.Rate'] == 118, 'Grad.Rate'] = 100 # fixing an outlier in df grad.rate from 118 to 100
college["Private"] = college["Private"].astype("category")


college["Total_Spending"] = college["Room.Board"] + college["Books"] + college["Expend"] + college["Personal"]



correlation = college["Total_Spending"].corr(college["Grad.Rate"])
print(f"Correlation between Total Spending and Graduation Rate: {correlation:.2f}")

plt.figure(figsize=(10, 6))
sns.scatterplot(data=college, x="Total_Spending", y="Grad.Rate", hue="Private", alpha=0.7)
plt.title("Total Spending vs. Graduation Rate")
plt.xscale("log")
# note on the log scale: 
# since we arent very interested in the anything above $30-40k (extremes cases)
# log scale will make the pattern clearer to see.
# (although, it might be percieved as misleading unless stated imo.)
plt.xlabel("Total Spending ($)")
plt.ylabel("Graduation Rate (%)")
plt.legend(title="Private College")
plt.show()
```

## Correlation between faculty quality and graduation rate
Lets look at faculty quality instead and see if it's more significant.

by combining the phd and terminal columns we can assume a new faculty quality column.

when plotting this against graduation rate and for some extra flavor coloring this with the S.F.Ratio data we can see a clear correlation between the faculty quality and graduation rate, and a also a correlation between a low Student/faculty ratio and graduation rate.

```python
college["faculty_quality"] = college["PhD"] + college["Terminal"]
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=college,
    x="faculty_quality",
    y="Grad.Rate",
    hue="S.F.Ratio",
    palette="Spectral",
    alpha=0.8
)
plt.xlim(0,)
plt.ylim(0,)

plt.title("Faculty Quality vs. Graduation rate")
plt.show()
```
